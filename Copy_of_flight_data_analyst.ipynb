```python
Y
#final assesement of 2nd quater


#loading flight dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('airlines_flights_data.csv')
df.head()
#checing null value
df.isnull().sum()
#there is no null value , now going for data pre _processign
#q .no .1 what are the airlines in the datasey accompanied by their frequencies
df['airline'].value_counts(ascending=True).plot.barh(color=['lightgreen','lightblue'])
plt.title('FREQUNCIES OF FLIGHT')
plt.xlabel('NUMBERS OF FLIGHT')
plt.show()
```

```python
#2 NO QUESTION SHOW THE BARGRAPH REPRESENTING THE DEPARTURE AND ARRIVAL TIME
df.head()
df['departure_time'].value_counts(ascending=True).plot.barh(color=['lightgreen','lightblue'])
plt.title('DEPARTURE TIME')
plt.xlabel('NUMBERS OF FLIGHT')
plt.show()
df['arrival_time'].value_counts(ascending=True).plot.barh(color=['lightgreen','lightblue'])
plt.title('ARRIVAL TIME')
plt.xlabel('NUMBERS OF FLIGHT')
plt.show()

```

```python
# Q.no 3 show bar graph representing the source city and destination city
df['source_city'].value_counts(ascending=True).plot.barh(color=['yellow','brown'])
plt.title('SOURCE CITY')
plt.xlabel('NUMBERS OF FLIGHT')
plt.show()
df['destination_city'].value_counts(ascending=True).plot.barh(color=['brown','lightblue'])
plt.title('DESTINATION CITY')
plt.xlabel('NUMBERS OF FLIGHT')
plt.show()
```

```python

# Q no 4 Does price varies with airline..?
df.head()
df.groupby('airline')['price'].mean()
#drawing the categorical plot showing the mean ticket of each airline
sns.catplot( x='airline' , y='price' , kind='bar' , data=df)
plt.title('MEAN TICKET PRICE')
plt.show()


```

```python
#Q NO 5 -- Does the ticket price changed based on the departure and arrival times
df.head()
df.groupby('departure_time')['price'].mean()
#drawing the categorical plot showing the mean ticket of each departure time
sns.catplot( x='departure_time' , y='price' , kind='bar' , data=df , palette='rocket',legend=False)
plt.title('MEAN TICKET PRICE')
plt.show()
df.groupby('arrival_time')['price'].mean()
#drawing the categorical plot showing the mean ticket of each arrival time
sns.catplot( x='arrival_time' , y='price' , kind='bar' , data=df , palette='rocket',legend=False)

```

```python
#lets see in line graph
df.head()
sns.relplot(x='arrival_time',y='price',data=df,col='departure_time',kind='line')
plt.show()

```

```python
# Q NO 6 how the price change with change in source and destination
df.head()
df.groupby('source_city')['price'].mean()
#drawing the categorical plot showing the mean ticket of each source city
sns.catplot( x='source_city' , y='price' , kind='bar' , data=df , palette='rocket',legend=False)
df.groupby('destination_city')['price'].mean()
#drawing the categorical plot showing the mean ticket of each destination city
sns.catplot( x='destination_city' , y='price' , kind='bar' , data=df , palette='rocket',legend=False)
```

```python
sns.relplot(x='source_city',y='price',data=df,col='destination_city',kind='line')
plt.show()
```

```python
# Q NO 7 how is the price affected when tickets are bought just before 1 0r 2 days departure
df.head()
df.days_left.unique()
df.groupby('days_left')['price'].mean()
#drawing the categorical plot showing the mean ticket of each days left
sns.relplot(x='days_left',y='price',data=df,kind='line')
plt.show()

```
